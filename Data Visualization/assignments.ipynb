```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(70, 10, 200)
sns.histplot(data, bins=20)
plt.axvline(x=70, color="red")
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()

sns.relplot(
    data=penguins, x="flipper_length_mm", y="body_mass_g", kind="scatter", hue="species"
)
```

```python
tips = sns.load_dataset("tips")

tips.head()

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker")

plt.xlabel("Total Bill")
plt.ylabel("Tips")
```

```python
tips_pivot = tips.pivot_table(
    index="day",
    columns="time",
    values="tip",
    aggfunc="mean"
)

# sns.heatmap(tips_pivot)
sns.heatmap(tips_pivot, annot = True, fmt = ".2f")
```
