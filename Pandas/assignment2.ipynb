```python
df_age = df[(df["age"] >= 25) & (df["age"] <= 35)]
df_age
```

```python
df["salary_after_tax"] = df["salary"] - (df["salary"] * 0.1)
df
```

```python
df["joining_date_time"] = pd.to_datetime(df["doj"])
df
```

```python
df_doj = df[(df["joining_date_time"] > "2021-01-01")]
df_doj
```

```python
city = df.groupby("city")["salary"].mean()
city

max_salary = df.groupby("city")["salary"].max()
max_salary

employee_count = df.groupby("city")["emp_id"].count()
employee_count
```

```python
# df.groupby('department')['salary'].mean()
df["salary"] = df["salary"].fillna(df.groupby("department")["salary"].transform("mean"))
df
```

```python
# import numpy as np
# today = pd.to_datetime('now')

# df['experience_years'] =  (today - df['joining_date_time']) / np.timedelta64(1 , 'Y')

today = pd.Timestamp.today()

df["experience_years"] = round((today - df["joining_date_time"]).dt.days / 365.25, 2)
df
```
