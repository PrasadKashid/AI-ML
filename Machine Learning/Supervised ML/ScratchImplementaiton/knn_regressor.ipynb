```python
import numpy as np
```

```python
class KNNRegressor:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def _euclidean_dist(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2, axis=1))

    def _predict_one(self, x):
        distance = self._euclidean_dist(self.X_train, x)
        knn_indices = np.argsort(distance)[: self.k]

        knn_values = self.y_train[knn_indices]

        return np.mean(knn_values)

    def predict(self, X):
        X = np.array(X)
        y_pred = [self._predict_one(x) for x in X]
        return np.array(y_pred)
```

```python
X_train = [[1], [2], [3], [4], [5]]
y_train = [2, 4, 6, 8, 10]

X_test = [[1.5], [3.5], [6]]
```

```python
model = KNNRegressor()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)
```
