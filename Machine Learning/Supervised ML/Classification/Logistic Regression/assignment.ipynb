```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
employee_data = pd.read_csv("../../../Datasets/employee_turnover.csv")
```

```python
employee_data.head()
employee_data.info()
```

```python
X = employee_data.drop("Employee_Turnover", axis=1)
y = employee_data["Employee_Turnover"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.2
)
```

```python
logReg = LogisticRegression(max_iter=200)
logReg.fit(X_train, y_train)
```

```python
y_pred = logReg.predict(X_test)
y_pred
```

```python
print("Accuracy", accuracy_score(y_pred=y_pred, y_true=y_test) * 100, "%")
print("Precision", classification_report(y_pred=y_pred, y_true=y_test) * 100, "%")
```

```python
# Lasso Model
lasso = LogisticRegression(penalty="l1", solver="liblinear", C=0.5)
lasso.fit(X_train, y_train)
```

```python
# Ridge Model l2
ridge = LogisticRegression(penalty="l2", C=1, max_iter=200)
ridge.fit(X_train, y_train)
```

```python
model = {"BaseLine": logReg, "lasso": lasso, "ridge": ridge}

for name, model in model.items():
    y_pred = model.predict(X_test)
    print(f"\n{name}")
    print("Accuracy Score: ", round(accuracy_score(y_test, y_pred) * 100, 2), "%")
    print(classification_report(y_test, y_pred))
```
