```python
import requests

URL = 'https://dogapi.dog/api/v2/breeds?page%5Bnumber%5D=1&page%5Bsize%5D=100'

res = requests.get(URL)

res
```

```python
print(res.headers.get("Content-Type"))

json_data = res.json()
```

```python
import pandas as pd

df = pd.json_normalize(json_data['data'])
df = df[['type', 'attributes.name','attributes.description','attributes.life.max']]

df = df.rename(
    columns={
        'attributes.name' : 'Name',
        'attributes.description' : 'Description',
        'attributes.life.max' : 'Max_life'
    }
)

df
```
