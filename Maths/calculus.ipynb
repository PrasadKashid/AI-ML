```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100_000)
y = np.sqrt(x)

plt.plot(x, y, label="y = sqrtx")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sqrt - x")
plt.legend()
plt.grid(True)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label="y = sin(x)", color="blue")

plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label = "x = 0")
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label = "y = 0")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = sin(x)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
x = np.linspace(-2 * np.pi / 2, 2 * np.pi / 2, 100)
y = np.sin(x)

plt.plot(x, y, label="y = sin(x)", color="blue")

plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label = "x = 0")
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label = "y = 0")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = sin(x)")
plt.grid(True)
plt.legend()

plt.show()
```
