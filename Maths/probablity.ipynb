```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(0, 10, 10_00_000)

plt.hist(values, bins=100, density=True, alpha=0.3)

plt.title("Continuous Uniform Distribution between 0 and 10")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 70
sigma = 10

x = np.linspace(30, 110, 1000)

y = norm.pdf(x, mu, sigma)

plt.plot(x, y, color="black", linewidth=2, label="Normal Dist")

x1 = np.linspace(mu - sigma, mu + sigma, 1000)
plt.fill_between(
    x1, norm.pdf(x1, mu, sigma), color="green", alpha=0.3, label="68% region"
)

x2 = np.linspace(mu - 2 * sigma, mu + 2 * sigma, 1000)
plt.fill_between(
    x2, norm.pdf(x2, mu, sigma), color="yellow", alpha=0.2, label="95% region"
)

x3 = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 1000)
plt.fill_between(
    x3, norm.pdf(x3, mu, sigma), color="red", alpha=0.1, label="99.7% region"
)

plt.title("Normal Dist")
plt.xlabel("Score")
plt.ylabel("Prob Density")
plt.legend()
plt.show()
```
